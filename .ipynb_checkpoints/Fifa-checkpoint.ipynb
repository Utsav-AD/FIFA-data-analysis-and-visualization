```python
#importing required libraries 
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#reading the csv file
fifa =  pd.read_csv('players_20.csv')
```

```python
#showing top 10 rows of the file
fifa.head()
```

```python
#checking the name of columns
for col in fifa.columns:
    print(col)
```

```python
#checking the number of rows and columns
fifa.shape
```

```python
#counting the number of players as per their nationality
fifa['nationality'].value_counts()
```

```python
#top 10 country with most number of players
fifa['nationality'].value_counts()[0:10]
```

```python
#bargraph showing the data of top 10 nations
plt.figure(figsize=(10, 6))
plt.bar(list(fifa['nationality'].value_counts()[0:10].keys()), list(fifa['nationality'].value_counts()[0:10]),color='royalblue', edgecolor='black')
plt.xlabel("Country", fontsize=12)
plt.ylabel("Number of Players", fontsize=12)
plt.title("Top 10 Nationalities in FIFA Dataset", fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#extracting the salary of player from the dataset
player_salary=fifa[['short_name','wage_eur']]
player_salary.head()
```

```python
#shorting it according to the highest wages
player_salary=player_salary.sort_values(by=['wage_eur'], ascending=False)
player_salary.head()
```

```python
#showing the data of players and their salary in bargraph
plt.figure(figsize=(10, 6))
bars = plt.bar(list(player_salary['short_name'][0:10]), list(player_salary['wage_eur'][0:10]), color='seagreen', edgecolor='black')
plt.xlabel("Player Name", fontsize=12)
plt.ylabel("Salary (EUR)", fontsize=12)
plt.title("Top 10 Players by Salary", fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f"{yval:,}", ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.show()
```
